### Data Preprocessing

##### Missing Value imputation

```python
data.drop(['LocationAbbr','Data_Value_Unit','Data_Value_Footnote_Symbol','Data_Value_Footnote','Total','Age(years)','Education','Gender','Income','Race/Ethnicity'] , axis=1, inplace=True)
data
```

#### Encoding

```python
data1=pd.get_dummies(data[['LocationDesc','DataValueTypeID','Datasource','Class','Topic','Question','ClassID','TopicID','Data_Value_Type','StratificationCategory1','Stratification1','StratificationCategoryId1','StratificationID1']], drop_first=True)
data1
```

```python
data2=pd.concat([data,data1], axis=1)
data2
```
