### Missing value imputation

```python
data.dropna(how='any', inplace=True)
data
```

### One hot encoding

```python
data1=pd.get_dummies(data['YT'], drop_first=True)
data1
```

```python
data_new=pd.concat([data, data1], axis=1)
data_new
```

```python
data_new.drop(['YT', 'OGC'], axis=1,inplace=True)
```

```python
data_new
```

### label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
columns_to_encode = ['OG','DC','Dept','Union','JF','Job']
```

```python
data_new[columns_to_encode] = data_new[columns_to_encode].apply(lambda x: le.fit_transform(x))
```

```python
data_new
```

### Separating X and Y variable

```python
X= data_new.drop('Total_Compensation', axis=1)
y=data_new[['Total_Compensation']]
```

```python
X
```

```python
y
```

### Performing train test split.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.25, random_state=0)
```

### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
```

```python
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

### Applying algo

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
```

```python

from sklearn.metrics import r2_score
r2=r2_score(y_test, y_pred)
r2
```

#### The accuracy of the algorithm is 98.7%
